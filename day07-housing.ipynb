# DATA Encoding

```python
df_cleaned = df.copy()
```

```python
df_cleaned
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.info()
```

```python
df_cleaned['mainroad'] = df_cleaned['mainroad'].map({"no":0, "yes":1})
df_cleaned['guestroom'] = df_cleaned['guestroom'].map({"no":0, "yes":1})
df_cleaned['basement'] = df_cleaned['basement'].map({"no":0, "yes":1})
df_cleaned['hotwaterheating'] = df_cleaned['hotwaterheating'].map({"no":0, "yes":1})
df_cleaned['airconditioning'] = df_cleaned['airconditioning'].map({"no":0, "yes":1})
df_cleaned['prefarea'] = df_cleaned['prefarea'].map({"no":0, "yes":1})
```

```python
df_cleaned.head()
```

```python
df_cleaned = pd.get_dummies(df_cleaned, columns=['furnishingstatus'])
```

# Model

```python
#Linear Regression
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_cleaned.columns
```

```python
X = df_cleaned.drop(['price'], axis = 1)
```

```python
y = df_cleaned['price']
```

```python
X.columns
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred1 = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, pred1)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, pred1)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, pred1)
```
